```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
player2021 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2021_per_game.html')
```

```
player2021[0]
```

```
player2021_df[['G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']] = player2021_df[['G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']].apply(lambda col:pd.to_numeric(col,errors='coerce'))
```
